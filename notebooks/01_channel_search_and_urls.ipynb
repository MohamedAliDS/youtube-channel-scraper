## Step 1: Load Channel Names

Read the input Excel file with channel names.

```python
# Read the Excel file with channel names
input_file = input("ðŸ“‚ Enter path to Excel file with channel names: ").strip().strip('"')

df = pd.read_excel(input_file)
channel_names = df['alias'].tolist()

print(f"ðŸ“Š Loaded {len(channel_names)} channel names")
print(f"\nFirst 5 channels:")
for name in channel_names[:5]:
    print(f" â€¢ {name}")
```

## Step 2: Initialize Scraper with 8 Workers

âš¡ **Performance: 8 parallel threads for faster searching**

```python
# Initialize scraper with 8 workers for faster channel search
# More workers = faster parallel processing
max_workers = 8
scraper = YouTubeScraper(max_workers=max_workers, headless=True)

print("âš™ï¸ YouTubeScraper initialized")
print(f"ðŸš€ Max workers: {max_workers} (FAST PARALLEL PROCESSING)")
print(f"â±ï¸ Expected time: ~2-5 minutes for 100 channels")
print(f"\nðŸ’¡ If you experience issues:")
print(f"   - Reduce workers to 4-6 if Chrome crashes")
print(f"   - Increase workers to 10-12 if your system is powerful")
```

## Step 3: Search for Channels

ðŸ” Searching YouTube for each channel in parallel...

```python
# Search for channels (parallelized with 8 workers)
print(f"ðŸ” Searching for {len(channel_names)} channels with {max_workers} parallel workers...\n")

results = scraper.search_channels(channel_names)

print(f"\nâœ… Search completed!")
print(f"ðŸ“Š Total channels processed: {len(results)}")

# Count found vs not found
found = sum(1 for r in results if r['channel_url'] != 'Not Found')
not_found = len(results) - found
print(f"âœ… Found: {found}")
print(f"âŒ Not found: {not_found}")
```

## Step 4: View Results

```python
# Display results
df_results = pd.DataFrame(results)

print("Sample Results (first 10 rows):")
print(df_results.head(10).to_string(index=False))
```

## Step 5: Save Results

```python
# Save results to Excel
output_file = "channels_found.xlsx"
output_path = Path(str(Path.cwd().parent / 'data' / 'processed' / output_file))

df_results.to_excel(output_path, index=False, engine="openpyxl")

print(f"âœ… Results saved to: {output_path}")
print(f"ðŸ“„ File: {output_file}")
print(f"ðŸ“Š Rows: {len(df_results)}")
```
